```python
import os
import pandas as pd

# Path to the folder containing the parquet files
parquet_folder = 'neumarkt_parquet'

# List all .parquet files in the folder
parquet_files = [f for f in os.listdir(parquet_folder) if f.endswith('.parquet')]

# Load each parquet file into a DataFrame and store in a dictionary
schedule = {}
for file in parquet_files:
    key = file.replace('.parquet', '')
    schedule[key] = pd.read_parquet(os.path.join(parquet_folder, file))

schedule["stops"] = schedule["stops"].rename(columns={"lat": "stop_lat", "lon": "stop_lon"})
```

```python
stops = schedule["stops"]
# Voggenthal Ort, Helena
skip_stops = ["de:09373:17126:0:1", "de:09373:17197:0:1"]

# ziegel h√ºttten
added_routes = {"r1": ["de:09373:17000:53:6", "de:09373:17108:0:1", "de:09373:17019:0:1", "de:09373:17006:0:1", "de:09373:17002:0:1", "de:09373:17000:53:6"]}
```

```python
import drawer

line_names = schedule["routes"]["route_short_name"].unique().tolist()
print(line_names)
lines_old = drawer.get_multiple_routes_positions(schedule, line_names)
line_names.remove("575")
line_names.remove("574")
lines_new = drawer.get_multiple_routes_positions(schedule, line_names, skip_stop_ids=skip_stops)
```
